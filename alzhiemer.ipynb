```python
def Multi_Layes_CNN_model(height, width):
    input_shape = (height, width, 1)
    print(input_shape)
    input = Input(shape=input_shape)
    x=Conv2D (10,(5,5),padding= 'same',activation= 'relu',kernel_regularizer=regularizers.l2(0.01))(input)
    print(x.shape)
   # x1 = Conv2D(5, (3, 3), padding='same', activation='relu',kernel_regularizer=regularizers.l2(0.01))(x)
    #x2 = Conv2D(5, (5, 5), padding='same', activation='relu',kernel_regularizer=regularizers.l2(0.01))(x)
    #x3 = Conv2D(5, (7, 7), padding='same', activation='relu',kernel_regularizer=regularizers.l2(0.01))(x)
    #x4 = Conv2D(50, (9, 9), padding='same', activation='relu')(x)
    #x=concatenate([x1,x2,x3],axis= 3)
   # x = BatchNormalization(momentum=0.95)(x)
    #x = Conv2D(1, (4, 4), padding='same', activation='relu')(x)
    x=MaxPooling2D ((2,2))(x)
    print(x.shape)
    x=Conv2D (20,(4,4),activation= 'relu',padding= 'same',kernel_regularizer=regularizers.l2(0.01))(x)
    print(x.shape)
    x=MaxPooling2D ((2,2))(x)
    print(x.shape)
    x=Conv2D(40,(3,3),activation= 'relu',padding= 'same',kernel_regularizer=regularizers.l2(0.01))(x)
    print(x.shape)
    x=MaxPooling2D ((2,2))(x)
    print(x.shape)
    x=Conv2D(50,(3,3),padding= 'same',activation= 'relu',kernel_regularizer=regularizers.l2(0.01))(x)
    print(x.shape)
    x=MaxPooling2D ((2,2))(x)
    print(x.shape)
    x=Conv2D(60,(2,2),activation= 'relu',padding= 'same',kernel_regularizer=regularizers.l2(0.01))(x)
    print(x.shape)
    x = MaxPooling2D((2, 2))(x)
    print(x.shape)
    x = Conv2D(80, (2, 2), activation='relu', padding='same',kernel_regularizer=regularizers.l2(0.01))(x)
    print(x.shape)
    x=Flatten ()(x)
  #  x = BatchNormalization(momentum=0.95)(x)
    print(x.shape)
    model = Model(input, x)
    return model

```

```python
def load_data(path):
    x = []
    y = []
    for image in os.listdir(path + '/MildDemented'):
        # print(j, image)
        img = cv2.imread(path + '/MildDemented/' + image, cv2.IMREAD_ANYDEPTH)
        x.append(img)
        y.append(0)
    for image in os.listdir(path + '/ModerateDemented'):
        # print(j, image)
        img = cv2.imread(path + '/ModerateDemented/' + image, cv2.IMREAD_ANYDEPTH)
        x.append(img)
        y.append(1)
    for image in os.listdir(path + '/VeryMildDemented'):
        # print(j, image)
        img = cv2.imread(path + '/VeryMildDemented/' + image, cv2.IMREAD_ANYDEPTH)
        x.append(img)
        y.append(2)
    for image in os.listdir(path + '/NonDemented'):
        # print(j, image)
        img = cv2.imread(path + '/NonDemented/' + image, cv2.IMREAD_ANYDEPTH)
        x.append(img)
        y.append(3)


    x = np.asarray(x)
    y = np.asarray(y)
    hot = LabelEncoder()
    y_encoded = hot.fit_transform(y)

    y_encoded = to_categorical(y_encoded, num_classes=4)
    return x, y_encoded
```
